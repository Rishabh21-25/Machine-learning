```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
company=soup.find_all('div',class_='companyCardWrapper__metaInformation')
```

```python
len(company)
```

```python
name=[]
rating=[]
reviews=[]
ctype=[]
hq=[]
old=[]
employees=[]
for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('span',class_='companyCardWrapper__companyRatingValue').text.strip())
    reviews.append(i.find('span',class_='companyCardWrapper__ActionTitle'))
    ctype.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip())
    hq.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip())
    old.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip())
    employees.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip())
d={'name':name, 'rating':rating, 'type':ctype,'hq':hq, 'old':old, 'eployees':employees}
df=pd.DataFrame(d)
df
```
